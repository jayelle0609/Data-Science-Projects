# ML evaluation tools 
> ## Classification

| **Metric**                   | **Use Case**                        | **Description**                                                      |
| ---------------------------- | ----------------------------------- | -------------------------------------------------------------------- |
| **Accuracy**                 | Balanced classes                    | Overall % of correct predictions: `(TP + TN) / (Total)`              |
| **Precision**                | Minimize false positives            | % of predicted positives that are actually correct: `TP / (TP + FP)` |
| **Recall (Sensitivity)**     | Minimize false negatives            | % of actual positives that are predicted correctly: `TP / (TP + FN)` |
| **F1 Score**                 | When precision & recall both matter | Harmonic mean of precision & recall: `2 * (P * R) / (P + R)`         |
| **Confusion Matrix**         | Classification breakdown            | Table showing TP, FP, FN, TN                                         |
| **ROC Curve**                | Visualize model performance         | Plots TPR vs FPR at various thresholds                               |
| **AUC-ROC**                  | Binary classification               | Area under ROC curve: 1 = perfect; 0.5 = random                      |
| **Log Loss / Cross-Entropy** | Probabilistic classifiers           | Penalizes confident but wrong predictions                            |


> ## Regression

| **Metric**                            | **Use Case**                           | **Description**                                        |
| ------------------------------------- | -------------------------------------- | ------------------------------------------------------ |
| **Mean Absolute Error (MAE)**         | Simple interpretation                  | Avg absolute difference: `mean(abs(y_true - y_pred))`  |
| **Mean Squared Error (MSE)**          | Penalize larger errors more            | Avg squared difference: `mean((y_true - y_pred)^2)`    |
| **Root Mean Squared Error (RMSE)**    | Similar to MSE, interpretable          | `sqrt(MSE)`  more intuitive units                     |
| **R (Coefficient of Determination)** | How well model explains variance       | Between 0 and 1 (closer to 1 = better)                 |
| **Adjusted R**                       | Multiple features                      | Corrects R for number of predictors                   |
| **MAPE**                              | When percentage error is more relevant | Mean Absolute Percentage Error: useful for time series |


> ## Clustering

| **Metric**                    | **Use Case**                | **Description**                                        |
| ----------------------------- | --------------------------- | ------------------------------------------------------ |
| **Silhouette Score**          | Evaluate clustering quality | Between -1 to 1 (higher = better cluster separation)   |
| **Inertia (SSE)**             | K-Means optimization        | Sum of squared distances to cluster centers            |
| **Davies-Bouldin Index**      | Internal cluster evaluation | Lower = better separation                              |
| **Adjusted Rand Index (ARI)** | When ground truth is known  | Measures similarity between clustering and true labels |
| **Calinski-Harabasz Score**   | Cluster compactness         | Higher = better defined clusters                       |

# Data Cleaning and Preprocessing
## 1. Clean Column Names : 
Remove extra spaces, replace special characters for easier access
## 2. Handle Missing Values : 
We will use a smart imputation strategy. Impute missing values for each country with its own average value, over the years, to preserve country-specific context. For any features that are still missing after this (ex. a country has no data for a feature), we will use the global median.

```python
# 1. Clean column names, removing whitespaces, replacing middle spaces and dashes with _
df.columns = (df.columns
              .str.strip()
              .str.replace(' ','_')
              .str.replace('-','_')
)

# Impute missing values
# Group by country and fill missing values with specific country's mean
for col in df.columns:
    if df[col].isnull().any() and df[col].dtype in [np.float64, np.int64]:
        # Use transform to apply the mean of each group back to the original df shape
        df[col] = df.groupby('Country')[col].transform(lambda x:x.fillna(x.mean()))

# For any remaining NAs (if a country has all missing values for a feature), fill with global median

df.fillna(df.median(numeric_only=True), inplace=True)

print("--- Missing Values after Cleaning ---")
print(df.isnull().sum())
```
